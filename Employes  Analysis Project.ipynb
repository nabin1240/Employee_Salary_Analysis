```python
# Phase 1 : Pandas Analysis
import pandas as pd
employee = {
    "Name" : ["Nabin", "Hari", "Lahas", "Krishna", "Govinda", "Sita", "Ram"],
    "Department" : [ "IT", "HR", "IT", "Finance", "HR", "IT", "Finance"],
    "Salary": [250000,400000,560000,433400,320000,500000,600000]
}
df = pd.DataFrame(employee)
print(df)
```

```python
#Analysis
print("Total Employees:", len(df))
```

```python
print("Average Salary :", df["Salary"].mean())
```

```python
print("Highest Salary:", df["Salary"].max())
```

```python
print("Lowest Salary:", df["Salary"].min())
```

```python
# GroupBy (pevot table)
print(df.groupby("Department")["Salary"].sum())
```

```python
#Groupby sum
print(df.groupby("Department")["Salary"].sum())
```

```python
#Groupby mean()
print(df.groupby("Department")["Salary"].mean())
```

```python
#Groupby  max
print(df.groupby("Department")["Salary"].max())
```

```python
#Groupby min
print(df.groupby("Department")["Salary"].min())
```

```python
# Groupby count
print(df.groupby("Department")["Salary"].count())
```

```python
# Highest Paid Employee
print(df.loc[df["Salary"].idxmax()])
```

```python
 # Ploting In chart
import matplotlib.pyplot as plt
dept_salary = df.groupby("Department")["Salary"].mean()
dept_salary.plot(kind="bar")
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Salary")
plt.show()
```

```python
#Phase 2 : SQL Version
import sqlite3
conn = sqlite3.connect("work.db")
cursor = conn.cursor()
cursor.execute(""" CREATE TABLE IF NOT EXISTS WORK (
NAME TEXT,
Department TEXT,
Salary INTEGER
)
""")
#INSERT DATA
cursor.execute("INSERT INTO WORK VALUES('Nabin', 'It', 25000)")
cursor.execute("INSERT INTO WORK VALUES('Hari', 'HR', 35000)")
cursor.execute("INSERT INTO WORK VALUES('Ram', 'IT', 50000)")
cursor.execute("INSERT INTO WORK VALUES('Sita', 'Finance', 70000)")
cursor.execute("INSERT INTO WORK VALUES('Rita', 'IT', 40000)")
cursor.execute("INSERT INTO WORK VALUES('Radha', 'Finance', 50000)")
cursor.execute("INSERT INTO WORK VALUES('HEMA', 'HR', 67000)")

conn.commit()
```

```python
# SQL Queries
#Highest Salary
cursor.execute("SELECT * FROM WORK ORDER BY Salary DESC  LIMIT 1;")
print(cursor.fetchall())
```

```python
# Average salary by department
cursor.execute("SELECT Department, AVG(salary)From WORK GROUP BY Department;")
print(cursor.fetchall())
```

```python
# Employees count(*)
cursor.execute("SELECT COUNT(*) FROM WORK")
print(cursor.fetchall())
conn.close()
```
